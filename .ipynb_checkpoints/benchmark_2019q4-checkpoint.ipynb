```python
def v_imps(row):
    if row['post_type'] == "Video":
        return row['impressions']
    else:
        return 0;

def v_thruplay_imps(row):
    if row['thruplay'] > 0:
        return row['impressions']
    else:
        return 0;
```

```python
Dato["post_type"]=Dato["video_views"].apply(lambda x: "Video" if x>0 else "Photo")
Dato["video_impressions"]=Dato.apply(lambda row: v_imps(row), axis=1)
Dato["video_thruplay_impressions"]=Dato.apply(lambda row: v_thruplay_imps(row), axis=1)
Dato = Dato[Dato['objective']!=""]
```

```python
global raw_data
global Dato_agg
group_by_val = list(['campaign_name', 'objective'])
Dato_agg = raw_data.groupby(group_by_val).sum()  
Dato_agg["VVR"] = Dato_agg["video_views"]/Dato_agg["video_impressions"]
Dato_agg["VCR"] = Dato_agg["video_fully_played"]/Dato_agg["video_impressions"]
Dato_agg["CTR"] = Dato_agg["website_link_clicks"]/Dato_agg["impressions"]
Dato_agg["SR"] = Dato_agg["shares"]/Dato_agg["impressions"]
Dato_agg["Engagement%"] = Dato_agg["total_engagements"]/Dato_agg["impressions"]
Dato_agg["VC25R"] = Dato_agg["video_complete_25%"]/Dato_agg["video_impressions"]
Dato_agg["VC50R"] = Dato_agg["video_complete_50%"]/Dato_agg["video_impressions"]
Dato_agg["VC75R"] = Dato_agg["video_complete_75%"]/Dato_agg["video_impressions"]
Dato_agg["Thruplay%"] = Dato_agg["thruplay"]/Dato_agg["video_thruplay_impressions"]
Dato_agg["CR"] = Dato_agg["comments"]/Dato_agg["impressions"]
Dato_agg["LR"] = Dato_agg["likes"]/Dato_agg["impressions"]
Dato_agg["CPM"] = Dato_agg["media_cost"]/(Dato_agg["impressions"]/1000)
Dato_agg["CP THRU"] = Dato_agg["media_cost"]/Dato_agg["thruplay"]
Dato_agg["CPV"] = Dato_agg["media_cost"]/Dato_agg["video_views"]
```

```python

def expl_handler(btn_object):
    #clear_output();
    global raw_data
    global Dato_agg
    global group_by_val
    global all_table
    global raw_data_bf_gp
    raw_data_bf_gp = raw_data
    
    group_by_val = list(groupby_dropdown.value)

    if 'campaign_name' not in group_by_val:
        group_by_val.append('campaign_name')
    print('group by->', group_by_val)
    if len(group_by_val) > 0:
        Dato_agg = raw_data_bf_gp.groupby(group_by_val).sum()
    else:
        Dato_agg = raw_data_bf_gp.sum()    
    Dato_agg["VVR"] = Dato_agg["video_views"]/Dato_agg["video_impressions"]
    Dato_agg["VCR"] = Dato_agg["video_fully_played"]/Dato_agg["video_impressions"]
    Dato_agg["CTR"] = Dato_agg["website_link_clicks"]/Dato_agg["impressions"]
    Dato_agg["SR"] = Dato_agg["shares"]/Dato_agg["impressions"]
    Dato_agg["Engagement%"] = Dato_agg["total_engagements"]/Dato_agg["impressions"]
    Dato_agg["VC25R"] = Dato_agg["video_complete_25%"]/Dato_agg["video_impressions"]
    Dato_agg["VC50R"] = Dato_agg["video_complete_50%"]/Dato_agg["video_impressions"]
    Dato_agg["VC75R"] = Dato_agg["video_complete_75%"]/Dato_agg["video_impressions"]
    Dato_agg["Thruplay%"] = Dato_agg["thruplay"]/Dato_agg["video_thruplay_impressions"]
    Dato_agg["CR"] = Dato_agg["comments"]/Dato_agg["impressions"]
    Dato_agg["LR"] = Dato_agg["likes"]/Dato_agg["impressions"]
    Dato_agg["CPM"] = Dato_agg["media_cost"]/(Dato_agg["impressions"]/1000)
    Dato_agg["CP THRU"] = Dato_agg["media_cost"]/Dato_agg["thruplay"]
    Dato_agg["CPV"] = Dato_agg["media_cost"]/Dato_agg["video_views"]
    
    Dato_brif = Dato_agg[['VVR', 'VCR', 'CTR', 'SR', 'Engagement%','VC25R', 'VC50R', 'VC75R', 'Thruplay%', 'CR', 'LR', 'CPM', 'CP THRU',
       'CPV']]
    
    group_by_val.remove('campaign_name')
    
    Dato_brif2 = Dato_brif.groupby(group_by_val).rank(ascending=True)
    Dato_brif3 = Dato_brif.groupby(group_by_val).count()
    Dato_bin = np.ceil(Dato_brif2/Dato_brif3*10)

    group_by_val.append('campaign_name')
    all_table = Dato_brif.merge(Dato_bin, left_index=True, on=group_by_val,suffixes=('', '_bin'),)
    pd.set_option('display.max_rows', all_table.shape[0]+1)
    display(all_table)

def expl_reset_handler(btn_object):
    clear_output();
    global raw_data_bf_gp
    raw_data_bf_gp = raw_data
    display(groupby_dropdown, btn_explo,btn_explo_reset)
```
